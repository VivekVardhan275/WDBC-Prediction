## Logistic Regression Model Training With Hyperparameter Tuning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
warnings.filterwarnings("ignore", category=ConvergenceWarning)
logistic_model = LogisticRegression()
param_grid = [
    {
        'penalty': ['l2'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['lbfgs', 'newton-cg', 'sag', 'saga'],
        'max_iter': [100, 500, 1000]
    },
    {
        'penalty': ['l1'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['lbfgs', 'newton-cg', 'sag', 'saga'],
        'max_iter': [100, 500, 1000]
    },
    {
        'penalty': ['elasticnet'],
        'l1_ratio': [0.1, 0.5, 0.9],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['lbfgs', 'newton-cg', 'sag', 'saga'],
        'max_iter': [100, 500, 1000]
    },
    {
        'penalty': ['none'],
        'solver': ['lbfgs', 'newton-cg', 'sag', 'saga'],
        'max_iter': [100, 500, 1000]
    }
]
grid_search_logistic = GridSearchCV(
    estimator=logistic_model,
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    verbose=1,
    n_jobs=-1
)
```

```python
import warnings
import os
os.environ["PYTHONWARNINGS"] = "ignore"
warnings.filterwarnings("ignore" , category=ConvergenceWarning)
warnings.filterwarnings("ignore" , category=DataConversionWarning)
grid_search_logistic.fit(X_train_scaled, y_train_scaled)
```

```python
y_predicted_logistic = grid_search_logistic.predict(X_test_scaled)
```

```python
print("Best Params:", grid_search_logistic.best_params_)
print("Best CV Accuracy:", grid_search_logistic.best_score_)

print("Test Accuracy:", accuracy_score(y_test_scaled, y_predicted_logistic))
print(classification_report(y_test_scaled, y_predicted_logistic))
print("Confusion Matrix:\n", confusion_matrix(y_test_scaled, y_predicted_logistic))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, auc
# Predict probability scores (for the positive class: Malignant = 1)
y_proba_logreg = grid_search_logistic.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test_scaled, y_proba_logreg)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='Logistic Regression (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Chance')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Logistic Regression')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

## Support Vector Machine Model Training With Hyperparameter Tuning

```python
from sklearn.svm import SVC
svc_model = SVC(probability=True)
param_grid_svc = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'gamma': ['scale', 'auto'],
    'degree': [2, 3, 4], 
    'shrinking': [True, False]
}
from sklearn.model_selection import GridSearchCV
grid_search_svc = GridSearchCV(
    estimator=svc_model,
    param_grid=param_grid_svc,
    scoring='accuracy',
    cv=5,
    verbose=1,
    n_jobs=-1
)
grid_search_svc.fit(X_train_scaled, y_train_scaled)
```

```python
print("Best Parameters:", grid_search_svc.best_params_)
print("Best CV Accuracy:", grid_search_svc.best_score_)

y_predicted_svc = grid_search_svc.predict(X_test_scaled)

print("Test Accuracy:", accuracy_score(y_test_scaled, y_predicted_svc))
print(classification_report(y_test_scaled, y_predicted_svc))
print("Confusion Matrix:\n", confusion_matrix(y_test_scaled, y_predicted_svc))
```

```python
y_proba_svc = grid_search_svc.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test_scaled, y_proba_svc)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='SVC (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Support Vector Classifier')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

## Random Forest Model Training With Hyperparameter Tuning

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
param_grid_rf = {
    'n_estimators': [100, 200],                 
    'max_depth': [3, 5, 10],                     
    'min_samples_split': [2, 5, 10],             
    'min_samples_leaf': [1, 2, 4],               
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True]    
}
grid_search_rf = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid_rf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)
grid_search_rf.fit(X_train_scaled, y_train_scaled)
```

```python
y_predicted_rf = grid_search_rf.predict(X_test_scaled)
print("Best Params:", grid_search_rf.best_params_)
print("Best Cross-Validation Accuracy:", grid_search_rf.best_score_)

print("Test Accuracy:", accuracy_score(y_test_scaled, y_predicted_rf))

print(classification_report(y_test_scaled, y_predicted_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test_scaled, y_predicted_rf))
```

```python
y_proba_rf = grid_search_rf.predict_proba(X_test_scaled)[:, 1]
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test_scaled, y_proba_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

plt.figure(figsize=(8, 6))
plt.plot(fpr_rf, tpr_rf, color='orange', lw=2, label='Random Forest (AUC = {:.2f})'.format(roc_auc_rf))
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Random Forest')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

## XGBoost Model Training With Hyperparameter Tuning

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(
    objective='binary:logistic',
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42,
    verbosity=0
)
param_grid_xg= {
    'n_estimators': [100, 200],         # Number of trees (boosting rounds)
    'max_depth': [3, 5, 7],             # Pre-pruning: max tree depth
    'learning_rate': [0.01, 0.1, 0.2],  # Step size shrinkage
    'subsample': [0.8, 1.0],            # % of rows used per tree
    'colsample_bytree': [0.8, 1.0],     # % of columns used per tree
    'gamma': [0, 0.1, 0.3],             # Pre-pruning: minimum loss reduction to make split
    'min_child_weight': [1, 3, 5]       # Pre-pruning: minimum sum of instance weight (hessian)
}
grid_search_xgb = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid_xg,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)
warnings.filterwarnings('ignore', message=".*use_label_encoder.*")
grid_search_xgb.fit(X_train_scaled, y_train_scaled)
```

```python
y_predicted_xgb = grid_search_xgb.predict(X_test_scaled)
print("Best Params:", grid_search_xgb.best_params_)
print("Best Cross-Validation Accuracy:", grid_search_xgb.best_score_)
print("Test Accuracy:", accuracy_score(y_test_scaled, y_predicted_xgb))
print(classification_report(y_test_scaled, y_predicted_xgb))
print("Confusion Matrix:\n", confusion_matrix(y_test_scaled, y_predicted_xgb))
```

```python
y_proba_xgb = grid_search_xgb.predict_proba(X_test_scaled)[:, 1]
fpr_xgb, tpr_xgb, thresholds_xgb = roc_curve(y_test_scaled, y_proba_xgb)
roc_auc_xgb = auc(fpr_xgb, tpr_xgb)
plt.figure(figsize=(8, 6))
plt.plot(fpr_xgb, tpr_xgb, color='purple', lw=2, label='XGBoost (AUC = {:.2f})'.format(roc_auc_xgb))
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])       
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - XGBoost')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

### ðŸ§¾ Conclusion

In this study, we conducted a comprehensive comparative analysis of various machine learning algorithms on the UCI Breast Cancer Wisconsin (Diagnostic) dataset. We evaluated Logistic Regression, Support Vector Machines (SVM), Naive Bayes, K-Nearest Neighbors (KNN), Random Forest, and XGBoost classifiers using grid search with cross-validation and tested them on a held-out test set.

Among all models, **XGBoost** demonstrated the **highest performance**, achieving a **test accuracy of 98.25%**, with strong precision, recall, and F1-scores across both classes. It also had the **lowest number of misclassifications**, making it the most reliable model in our evaluation. 

While Logistic Regression, SVM, and Random Forest also performed very well (each achieving ~97.37% accuracy), XGBoost consistently outperformed them in key metrics. 

This result confirms the effectiveness of ensemble boosting methods like XGBoost in handling structured medical datasets. Future work may include experimenting with feature engineering, ensemble stacking, and real-world deployment considerations for clinical use.
